```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import imageio
```

## Import data

```python
fert = pd.read_csv('data/gapminder_total_fertility.csv', index_col=0)
fert.head()
```

```python
life = pd.read_excel('data/gapminder_lifeexpectancy.xlsx', index_col=0)
life.head()
```

```python
pop = pd.read_excel('data/gapminder_population.xlsx', index_col=0)
pop.head()
```

```python
cont = pd.read_csv('data/continents.csv', sep=';')
cont.head()
```

## Data Wrangling

```python
print(fert.shape, life.shape, pop.shape, cont.shape)
```

```python
life.columns.dtype, fert.columns.dtype, pop.columns.dtype
```

```python
# converts columns (years) to integers
fert.columns = fert.columns.astype(int)
life.columns.dtype, fert.columns.dtype
```

```python
fert.index.name = 'country'
fert = fert.reset_index()
fert.head()
```

```python
# melt fert from wide to long format
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
fert.head()
```

```python
life.index.name = 'country'
life = life.reset_index()
life.head()
```

```python
# melt life from wide to long format
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
life.head()
```

```python
pop.index.name = 'country'
pop = pop.reset_index()
pop.head()
```

```python
# melt pop from wide to long format
pop = pop.melt(id_vars='country', var_name='year', value_name='population')
pop.head()
```

```python
# merge tables
df = fert.merge(pop)
df = df.merge(life)
df = df.merge(cont)
df.head()
```

```python
# drop missing values
df.dropna(inplace=True)
```

```python
df
```

## Define df_sample

```python
#chose a sample of df for the years between 1960 and 2015
df_sample = df[df['year'].between(1960, 2015)]
df_sample
```

## Plotting

```python
sns.set()
```

```python
# label and color specs
color_dict={'Africa': 'chocolate',
   'Asia': 'gold',
   'Europe': 'pink',
   'North America': 'lightcoral',
   'South America': 'seagreen',
   'Australia and Oceania': 'aqua'
           }
patches = [plt.plot([],[], marker="o", ms=20, ls="", mec=v, color=v, alpha=0.4,
            label=k)[0] for k,v in color_dict.items()]
plt.close()
```

```python
# Test for year 2015
fig = plt.figure(figsize=(20,10), dpi=100, facecolor='black')
plt.axes(fc="black")
plt.tick_params(axis='both', labelsize=15, labelcolor='darkgray')
plt.grid(c="darkgray")
plt.axis((10, 90, 0, 10))
plt.ylabel('Fertility rate', fontdict={'size': 20, 'color':'darkgray', 'weight': 'bold'})
plt.xlabel('Life expectancy', fontdict={'size': 20, 'color':'darkgray', 'weight': 'bold'})
s = df_sample[df_sample['year'] == 2015]['population']/20000
x = df_sample[df_sample['year'] == 2015]['life_expectancy']
y = df_sample[df_sample['year'] == 2015]['fertility_rate']
c = df_sample[df_sample['year'] == 2015].continent.replace(color_dict)
plt.scatter(x,y, s=s, c=c, alpha=0.4)
plt.title(label=str(2015), fontdict={'size': 30, 'color':'gray', 'weight': 'heavy'})
legend = plt.figlegend(handles=patches,
                       frameon=False, bbox_to_anchor=(0.9, 0.5), loc='center left',
                       labelspacing=1.5,
                       facecolor='black',
                       labelcolor='darkgray',
                       fontsize=20)
plt.savefig('./plots/test_'+str(2015)+'.png',
            dpi=60,
            bbox_extra_artists=(legend,),
            bbox_inches='tight')
plt.close(fig)
#plt.clf()
```

```python
# Plotting loop for all years
for i in range(1960,2016):
    plt.figure(figsize=(20,10), dpi=100, facecolor='black')
    plt.axes(fc="black")
    plt.tick_params(axis='both', labelsize=15, labelcolor='darkgray')
    plt.grid(c="darkgray")
    plt.axis((10, 90, 0, 10))
    plt.ylabel('Fertility rate', fontdict={'size': 20, 'color':'darkgray', 'weight': 'bold'})
    plt.xlabel('Life expectancy', fontdict={'size': 20, 'color':'darkgray', 'weight': 'bold'})
    color_dict={'Africa': 'chocolate',
       'Asia': 'gold',
       'Europe': 'pink',
       'North America': 'lightcoral',
       'South America': 'seagreen',
       'Australia and Oceania': 'aqua'
               }
    s = df_sample[df_sample['year'] == i]['population']/20000
    x = df_sample[df_sample['year'] == i]['life_expectancy']
    y = df_sample[df_sample['year'] == i]['fertility_rate']
    c = df_sample[df_sample['year'] == i].continent.replace(color_dict)
    plt.scatter(x,y, s=s, c=c, alpha=0.4)
    plt.title(label=str(i), fontdict={'size': 30, 'color':'gray', 'weight': 'heavy'})
    legend = plt.figlegend(handles=patches,
                           frameon=False,
                           bbox_to_anchor=(0.9, 0.5),
                           loc='center left',
                           labelspacing=1.5,
                           facecolor='black',
                           labelcolor='darkgray',
                           fontsize=20)
    plt.savefig('./plots/lifeexp_fert_'+str(i)+'.png',
                dpi=60,
                bbox_extra_artists=(legend,),
                bbox_inches='tight')
    plt.close()
    plt.clf()
```

```python
# output gif
images = []

for i in range(1960, 2016):
    filename = './plots/lifeexp_fert_'+str(i)+'.png'
    images.append(imageio.imread(filename))

imageio.mimsave('./lifeexp_fert_1960-2015.gif', images, fps=7)
```
